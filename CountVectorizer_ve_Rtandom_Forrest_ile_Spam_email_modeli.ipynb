```python
from datasets import load_dataset
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
dataset = load_dataset("anilguven/turkish_spam_email")
```

```python
type(dataset)
```

```python
train_df = dataset['train'].to_pandas()
test_df = dataset['test'].to_pandas()
```

```python
train_df.head()
```

```python
train_df = train_df.drop(['Unnamed: 0'], axis=1)
test_df = test_df.drop(['Unnamed: 0'], axis=1)
```

```python
train_df.head()
```

*Label 1 = spam e mail*
*Label 2 = not spam e mail*

# Data'dan stopwords'leri temizlemek

```python
import nltk
from nltk import download
download('stopwords')
from nltk.corpus import stopwords
```

```python
turkce_stop_words = stopwords.words('turkish')
```

```python
type(turkce_stop_words)
```

```python
for word in turkce_stop_words:
  word = " "+word+" "
  train_df['text'] = train_df['text'].str.replace(word, " ")
  test_df['text'] = test_df['text'].str.replace(word, " ")
```

```python
train_df["text"][0]
```

# CountVectorizer'ı çağırmak

```python
vectorizer = CountVectorizer(max_features=300)
x_train = vectorizer.fit_transform(train_df['text']).toarray()
y_train = train_df['labels']

x_test = vectorizer.transform(test_df['text']).toarray()
y_test = test_df['labels']
```

# Classifier modeli kurmak

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
model = rf.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
# Veri setini detaylıca incelediğinizde çok fazla yazım hatası olmakla birlikte eksik ve anlamsız cümle de olduğu için bu score gayet iyi görünüyor. Dateset sadece işlem sırası ve sınıflandırması için test amaçlı seçilmiştir.
```

# Modeli kullanmak

```python
new_email = ["yılbaşı çekilişleri var"]

for word in turkce_stop_words:
    new_email[0] = new_email[0].replace(" " + word + " ", " ")

new_email_vectorized = vectorizer.transform(new_email)


prediction = model.predict(new_email_vectorized)

if prediction == 1:
    print("Bu bir spam e-posta.")
else:
    print("Bu bir ham e-posta.")
```
